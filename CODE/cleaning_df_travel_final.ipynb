```python
df_travel = pd.read_csv("../DATA/TRAVEL/mrt_df_final.csv")
df_travel
```

```python

for i in range(len(df_travel)):
    
    if "유심" in df_travel['title'][i]:
        df_travel['drop'][i] = "Y"
    elif "픽업" in df_travel['title'][i]:
        df_travel['drop'][i] = "Y"
    elif "와이파이" in df_travel['title'][i]:
        df_travel['drop'][i] = "Y"
    else:
        df_travel['drop'][i] = "N"
```

```python
df_travel_clean = df_travel[df_travel['drop'] == "N"].reset_index()
```

```python
df_travel_clean = df_travel_clean.drop('index', axis=1)
```

```python
df_travel_clean2 = df_travel_clean[df_travel_clean['score'] >= 3.0].reset_index()
df_travel_clean3 = df_travel_clean2.drop('index', axis=1)
df_travel_clean3
```

```python
df_travel_clean3.to_csv("../DATA/TRAVEL/clean_mrt_df_final.csv", index=False)
```
