## Project Summary
### What is Broomie?
- `The best tour recommendation system for reasers like the best friends`
- Sounds "Brooreung~" in Korean Onomatopoeia: the sound when a car or an airplane starts(Vroom!)
- Means 'A combination of the four following words: besties, bros, homies, and roomies'
- Wish "Broomie" becomes the best friend for readers recommending the best tour based on their reading type 

## Modeling Plan
### System Structure
- Define the best tour type using MultilabelNB from 2-3 type possibilities
- Study tour_type using MultinomialNB(define tour type among 9 entities)  
    - Nature(자연친화)
    - City(도시)
    - Knowledge(지식추구)
    - Culture&Art(문화예술)
    - Challenge(도전)
    - Familiy(가족)
    - Experience(체험중시)
    - Selfie(셀피)
    - Food-holic(맛집)
- Make corpus by tour type
- Study detailed tour_type using MultinomialNB2
    - Nature(자연친화) : seeing stars, the seaside, plant-holic, animal-holic
    - City(도시) : seeing nightlight, walking local street, interesting fashion 
    - Knowledge(지식추구) : history, science
    - Culture&Art(문화예술) : music, museum, art museum 
    - Challenge(도전) :  hiking, extreme sports, 
    - Familiy(가족) : with lover, with parents, with children 
    - Experience(체험중시) : writing, making stuffs
    - Selfie(셀피) : snap picture, 
    - Food-holic(맛집) : cooking, local food, Michelin et Cie
- link book item to tour based on type and detailed tour type

### Data Introduction
- Book Data : 
    - site : Kyobo
    - columns :ISBN, title, author, book summary, book review, book keyword
    - total : 1507 items, 20 category(genres)
- Travel Data : 
    - site : myrealtrip
    - columns : tour_code, tour_name, tour_info, tour_review
    - total : 71 tour itmes, 8 cities
    
### Process
- Labeling : 9 tour types and detailed types to total tour data(71) & 1000 book data
- Use MultilabelNB, define the best type
- Use MultinomialNB, Study labels from 700 datats, test 300datas, predict 500 datats
- Input book title , Output best tour title and infos
